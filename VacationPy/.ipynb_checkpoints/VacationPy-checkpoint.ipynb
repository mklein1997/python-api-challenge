### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
locations = weatherdata[["Latitude", "Longitude"]].astype(float)

humidity = weatherdata['Humidity'].astype(float)

fig = gmaps.figure()

heatlayer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=2)
fig.add_layer(heatlayer)

fig
```
